# Initial Values and Population Generation for Knapsack Problem

In this cell, we define functions to set up initial values and generate a population for a knapsack problem. Here's a breakdown of what each function does:

1. **initial_values()**:
   - Initializes the weight limit `W` and arrays `w` and `v` representing weights and values of items respectively.
   - Computes the value-to-weight ratio and sorts the items based on this ratio.
   - Returns the sorted weights, values, and weight limit.

2. **value_generation(groups, individuals)**:
   - Calls `initial_values()` to get sorted weights and values.
   - Generates a population matrix with uniformly distributed random values.
   - The generated individuals are divided into specified groups.
   - Returns the population array.

3. **weight_generation(num)**:
   - Generates an array of random weights, normalizes them to sum to 1, and sorts them in descending order.
   - Returns the sorted weight array.


```python
def initial_values():
    W = 1433
    w = np.array([24, 45, 15, 40, 9, 37, 13, 5, 43, 35, 48, 50, 27, 46, 24, 45, 2, 7, 38, 20, 20, 31, 2, 20, 3, 35, 27, 4, 21, 22, 33, 11, 5, 24, 37, 31, 46, 13, 12, 12, 41, 36, 44,36, 34, 22, 29, 50, 48, 17, 8, 21, 28, 2, 44, 45, 25, 11, 37, 35, 24, 9, 40, 45, 8, 47, 1, 22, 1, 12, 36, 35, 14, 17, 5]) 
    v= np.array([2, 73, 82, 12, 49, 35, 78, 29, 83, 18, 87, 93, 20, 6, 55, 1, 83, 91, 71, 25, 59, 94, 90, 61, 80, 84, 57, 1, 26, 44, 44, 88, 7, 34, 18, 25, 73, 29, 24, 14, 23, 82, 38, 67, 94, 43, 61, 97, 37, 67, 32, 89, 30, 30, 91, 50, 21, 3, 18, 31, 97, 79, 68, 85, 43, 71, 49, 83, 44, 86, 1, 100, 28, 4, 16])
    res=v/w
    sorted_indices = np.argsort(res)
    #sorted_indices = np.argsort(v)
    w=w[sorted_indices]
    v=v[sorted_indices]
    return w,v,W

def value_generation(groups,individuals):
    w,v,W=initial_values()
    generated_individuals= np.random.uniform(0, 100, (groups * individuals, w.size))
    generated_individuals = np.round(generated_individuals, 2)
    population= np.array_split(generated_individuals, groups)
    return(population)

def weight_generation(num):
    w=np.random.random(num)
    return np.sort(w/w.sum())[::-1]
```

# Fitness Function Explanation

This function, `fitness_function`, is designed to calculate the fitness of a given solution in a genetic algorithm optimization problem. Let's break down its functionality:

1. **Input Parameters**: 
   - `l`: This parameter represents the solution being evaluated. It's a list of elements where each element denotes the proportion of an item selected in the solution.
   - `s`: This is an optional parameter (default is `False`). If set to `True`, the function returns additional information about the solution.

2. **Initialization**: 
   - It begins by obtaining the initial values of `w` (weights), `v` (values/profits), and `W` (maximum weight constraint) from the `initial_values()` function.

3. **Conversion of Proportions**: 
   - Each element in the solution `l` is converted into a fraction by dividing it by 100 and rounding to the nearest integer.

4. **Weighted Sums**: 
   - The weighted sums of weights (`wl`) and profits (`vl`) are computed by element-wise multiplication of the proportions with the corresponding weights and profits.

5. **Constraint Handling**: 
   - If the sum of weights exceeds the maximum weight constraint `W`, the function iterates through the solution and sets the proportion of items to 0 until the constraint is satisfied.

6. **Final Calculations**: 
   - After satisfying the weight constraint, the function computes the final profit by summing up the profits of the selected items.

7. **Output**: 
   - If the optional parameter `s` is set to `True`, the function returns a tuple containing the final profit and the updated solution proportions. Otherwise, it returns only the final profit.

This function is essential in genetic algorithms for evaluating the fitness of candidate solutions and guiding the search towards better solutions over successive generations.



```python
def fitness_function(l,s=False):
    w,v,W=initial_values()
    counter=0
    l1=[float(round(el/100)) for el in l]
    wl=w*l1
    vl=v*l1
    sum_of_weights=np.sum(wl)
    sum_of_profits=np.sum(vl)
    i=0

    while sum_of_weights>W:
        sum_of_profits=sum_of_profits-vl[i]
        sum_of_weights=sum_of_weights-wl[i]
        l1[i]= 0
        i+=1

    l2=[l1[i]*l[i] for i in range(len(l))]

    if s==True:
        return sum_of_profits,l2
    else:
        return sum_of_profits
```

# Sorting Function Explanation

The `sorting` function is responsible for sorting a list of solutions based on their fitness values. Let's break down its functionality:

1. **Input Parameter**: 
   - `L`: This parameter represents a list of solutions, where each solution is a list of elements.

2. **Global Best Initialization**:
   - It initializes `GLr` (Global Best Result) to 0. This variable will hold the highest fitness value encountered across all solutions.

3. **Iterating Through Solutions**:
   - It iterates through each sublist in the input list `L`.

4. **Computing Fitness and Global Best Update**:
   - For each element in the current sublist, it computes the fitness using the `fitness_function` and stores the result.
   - If the computed fitness for the current element is greater than the current `GLr` (Global Best Result), it updates `GLr` and captures the corresponding solution in `GL`.

5. **Sorting**:
   - It combines each solution element with its corresponding fitness value using `zip` and sorts them based on fitness values in descending order.
   - The sorted solutions are extracted and stored in `L[sublist]`.

6. **Output**:
   - After sorting all sublists, the function returns the sorted list of solutions `L` along with the global best solution `GL`.

This function is crucial in evolutionary algorithms, especially genetic algorithms, for arranging candidate solutions in descending order of fitness, facilitating the selection of better solutions for reproduction and creating the next generation.



```python
def sorting(L):
    GLr=0
    for sublist in range(len(L)):
        result=[]
        for element in range(len(L[0])):
            result.append(fitness_function(L[sublist][element]))
            if result[-1]>GLr:
                GLr=result[-1]
                GL=L[sublist][element]
        combined_data = list(zip(L[sublist][::], result))
        sorted_data = sorted(combined_data, key=lambda x: x[1], reverse=True)
        sorted_individuals = [item[0] for item in sorted_data]
        L[sublist]=np.array(sorted_individuals)
        
    return L,GL
```

# Update Search Function Explanation

The `update_search` function is responsible for updating the search space based on the global leader and the population. Let's break down its functionality:

1. **Input Parameters**: 
   - `population`: This parameter represents the current population of solutions.
   - `GL`: The global leader solution.

2. **Temporary Population Initialization**: 
   - It creates a deep copy of the population called `temp`.

3. **Updating Leaders and Advocates**: 
   - For each solution in the population:
     - It generates random weights `w1` and `w2` using the `weight_generation` function.
     - For the leader (`temp[i][0]`):
       - It computes the fitness of the leader and updates it based on a combination of the global leader, the advocate, and the mean of the remaining solutions in the population.
     - For the advocate (`temp[i][1]`):
       - It computes the fitness of the advocate and updates it based on a combination of the current leader and the mean of the remaining solutions in the population.

4. **Updating Believers**:
   - For the remaining solutions (`temp[i][2:]`):
     - It generates random weights `w3`.
     - It updates each believer based on a combination of the current leader and the advocate.

5. **Population Update**:
   - It assigns the updated solutions from `temp` back to `population`.

6. **Output**:
   - The function returns the updated population.

This function essentially promotes diversity and exploration in the search space by updating solutions towards the global leader while maintaining the diversity within the population.

```python
def update_search(population,GL):
    temp=[element.copy() for element in population]
    
    for i in range(len(temp)):
        w1=weight_generation(3)
        temp[i][0]=fitness_function(temp[i][0],True)[1]
        temp[i][0]=GL*w1[0]+population[i][1]*w1[1]+w1[2]*np.mean(population[i][2:],axis=0)

        w2=weight_generation(2)
        temp[i][1]==fitness_function(temp[i][1],True)[1]
        temp[i][1]=w2[0]*population[i][0]+w2[1]*np.mean(population[i][2:],axis=0)
        
        for j in range(2,len(temp[0])):
            w3=weight_generation(2)
            temp[i][j]=w3[0]*population[i][0]+w3[1]*population[i][1]

    population=[element.copy() for element in temp]
    
    return population
```

# Function Descriptions

## 1. `saturation`

The `saturation` function checks if the global leader values have converged within a given tolerance.

### Parameters:
- `GLl` (list): A list of global leader values.
- `tolerance` (float, optional): The tolerance for checking convergence. Default is `1e-6`.

### Returns:
- `bool`: `True` if the global leader values have converged within the specified tolerance, otherwise `False`.

## 2. `changeGL`

The `changeGL` function modifies the global leader values if they fall below a certain threshold. Priority is given to the start of the list for replacement.

### Parameters:
- `GL` (list): A list of global leader values.
- `threshold` (float, optional): The threshold value below which the global leader value will be replaced. Default is `50`.
- `replacement_value` (float, optional): The value to replace the global leader value with. Default is `51.0`.
- `priority_weight` (int, optional): The weight given to prioritizing indices for replacement. Default is `2`.

### Returns:
- `list`: The modified list of global leader values.

## 3. `iterations`

The `iterations` function performs multiple iterations of updating and sorting the population while tracking the global leader and the results of the objective function.

### Parameters:
- `num` (int): The number of iterations to perform.
- `data` (DataFrame): A DataFrame to store the results.
- `population` (list): The current population of solutions.
- `GL` (list): The global leader solution.
- `GL1` (list): A list to store the global leader's objective function values.
- `l` (int, optional): The starting iteration. Default is `0`.

### Returns:
- `data` (DataFrame): The updated DataFrame with iteration results.
- `population` (list): The updated population.
- `GL` (list): The updated global leader solution.
- `GL1` (list): The updated list of global leader's objective function values.


```python
def saturation(GLl, tolerance=1e-6):
    return abs(sum(GLl)/len(GLl) - GLl[-1]) < tolerance
```

```python
def changeGL(GL, threshold=50, replacement_value=51.0, priority_weight=2):
    
    # Create a list of indices with priority given to the start
    priority_indices = list(range(len(GL)))
    priority_indices.extend([random.choice(range(len(GL))) for _ in range(priority_weight)])

    # Shuffle the list to randomize the selection
    random.shuffle(priority_indices)
    count=0
    for index in priority_indices:
        #print(GL[index] < threshold)
        if GL[index] < threshold:
            GL[index] = replacement_value
            #print(count)
            count=count+1
        if count==2:
            break
    return GL
```

```python
def iterations(num,data,population,GL,GL1,l=0):
    for i in range(l,num):
        population=update_search(population,GL)
        population,GL=sorting(population)
        
        # Flatten the nested list
        flat_list = [item for sublist in population for item in sublist]

        # Storing the results of the objective function
        A = [fitness_function(i,True)[0] for i in flat_list]
        B = [fitness_function(i,True)[1] for i in flat_list]

        new_data = {'Iteration Number': i+1, 'Values':flat_list, 'Result': A,'Values Used':B}
        new_df = pd.DataFrame(new_data)

        # Create a DataFrame with iteration number and values
        data = pd.concat([data, new_df], ignore_index=True)
   
        A=fitness_function(GL)
        GL1.append(A)
        if i%10==0:
            if saturation(GL1):
                nGL=changeGL(GL)
                run=0
                while fitness_function(nGL)<= fitness_function(GL):
                        nGL=changeGL(GL)
                        run=run+1
                        if run==5:
                             break
                GL=nGL
                GL1=[]
            else:
                GL1=[]

    return data,population,GL,GL1
```
